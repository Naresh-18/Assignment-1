```python
def reindex_dataframe(df):
    new_index = range(1, len(df)+1, 2) # create new index starting from 1 and incrementing by 2
    df = df.reindex(new_index) # re-index the DataFrame with the new index
    return df
```

```python
# create sample DataFrame
df = pd.DataFrame({'A': ['X', 'Y', 'Z', 'A','B','C'], 'B': [True, False,True, True, True, False], 'C': [1, 2, 3, 4, 5, 6]})
df
```

```python
# re-index the DataFrame
df_reindexed = reindex_dataframe(df)
df_reindexed
```

```python
'''In pandas, DataFrame.size and DataFrame.shape are two different attributes that provide different information about a DataFrame.

1.DataFrame.size:

 The size attribute returns the total number of elements in the DataFrame.
 It is equivalent to the total number of cells in the DataFrame, 
 calculated by multiplying the number of rows by the number of columns.
 It returns a single integer value.'''
import pandas as pd

# Creating a sample DataFrame
data = {'A': [1, 2, 3], 'B': ['apple', 'banana', 'orange']}
df = pd.DataFrame(data)
total_size = df.size
print(f"Total size of DataFrame: {total_size}")
```

```python
'''
2.DataFrame.shape:

 The shape attribute returns a tuple representing the dimensions of the DataFrame.
 The tuple consists of two elements: the number of rows and the number of columns.
 It provides information about the structure of the DataFrame.'''
import pandas as pd

# Creating a sample DataFrame
data = {'A': [1, 2, 3], 'B': ['apple', 'banana', 'orange']}
df = pd.DataFrame(data)
dimensions = df.shape
print(f"Number of rows: {dimensions[0]}, Number of columns: {dimensions[1]}")
```

```python
def select_rows(df):
    # create a boolean mask for rows where 'A' > 5 and 'B' < 10
    mask = (df['A'] > 5) & (df['B'] < 10)
    
    df_new = df[mask].copy()
    return df_new
df = pd.DataFrame({'A':[3,8,6,2,9],'B':[5,2,9,3,1],'C':[1,7,4,5,2]})
df
```

```python
df_new = select_rows(df)
df_new
```

```python
def calculate_stats(df):
    # calculate the mean, median, and standard deviation of the 'Values' column
    mean = df['Values'].mean()
    median = df['Values'].median()
    std_dev = df['Values'].std()

    return mean, median, std_dev
```

```python
df = pd.DataFrame({'Values': [18, 7, 1, 45, 17]})
df
```

```python
mean, median , stdev = calculate_stats(df)
print(f'Mean of Values is : {mean}')
print(f'Median of Values is : {median}')
print(f'Standard Deviation of Values is : {round(stdev,4)}')
```

```python
import pandas as pd

def calculate_moving_average(df):
    # sort the DataFrame by date
    df = df.sort_values(by='Date')

    # calculate the moving average using a rolling window of size 7
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()

    return df
```

```python
df = pd.DataFrame({
    'Date': ['2021-10-14', '2021-10-15', '2021-10-16', '2021-10-17', '2021-10-18', '2021-10-19', '2021-10-20', '2021-10-21', '2021-10-22', '2021-10-23', '2021-10-24', '2021-10-25'],
    'Sales': [20, 12, 45, 18, 42, 64, 32, 21, 55, 81, 75, 113]
})

df = calculate_moving_average(df)
```

```python
df_new = calculate_moving_average(df)
df_new
```

```python
def return_weekdays(df):

    # Below changes String Date to datetime format
    df['Date']=pd.to_datetime(df['Date'])
    df['Weekday'] = df['Date'].dt.day_name()
    return df
```

```python
df = pd.DataFrame({'Date':['2022-12-03','2022-12-04','2022-12-05','2022-12-06','2022-12-07']})
df
```

```python
df_new = return_weekdays(df)
df_new
```

```python
def select_December_rows(df):
    # Convert 'Date' column to pandas datetime format
    df['Date'] = pd.to_datetime(df['Date'])
    December_rows = df[df['Date'].between('2022-12-01', '2022-12-31')]
    return December_rows
```

```python
df = pd.DataFrame({'Date':pd.date_range(start='1/1/2022',end='1/1/2023',freq='D')})
df
```

```python
select_December_rows(df)
```
