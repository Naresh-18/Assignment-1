The three measures of central tendency are:


   1.Mean: The mean, often referred to as the average, is calculated by adding up all the values in a data set and            then dividing the sum by the number of values. 

   2.Median: The median is the middle value in a sorted list of numbers. If the data set has an odd number of                    observations, the median is the middle value. If the data set has an even number of observations, the              median is the average of the two middle values.

   3.Mode: The mode is the value that appears most frequently in a data set. A data set may have one mode, more                than one mode, or no mode at all.

# Question 5

```python
Measures of dispersion, including range, variance, and standard deviation, are used to quantify the extent to which individual data points in a dataset deviate from the central tendency (mean, median, etc.). 
They provide insights into the variability or spread of the data points. 
Here's a brief explanation of each measure:

Range:

- Definition: The range is the simplest measure of dispersion and is calculated as the difference between the maximum and minimum values in a dataset.
- Example: Consider two datasets:
           Dataset A: [10, 15, 20, 25, 30]
           Dataset B: [5, 15, 25, 35, 45]
  The range for Dataset A is 30 - 10 = 20, and for Dataset B it is 45 - 5 = 40.

Variance:

- Definition: Variance is the average of the squared differences between each data point and the mean of the dataset.
- Example: Let's take a dataset:
           [12, 15, 18, 22, 25]
  First, find the mean: (12 + 15 + 18 + 22 + 25) / 5 = 18.4
  Then, calculate the squared differences from the mean: [(12-18.4)^2, (15-18.4)^2, (18-18.4)^2, (22-18.4)^2, (25-18.4)^2]
  The variance is the average of these squared differences.
Standard Deviation:

- Definition: Standard deviation is the square root of the variance. 
              It provides a measure of the average distance between each data point and the mean.
- Example: Continuing from the previous example, once you find the variance, take the square root to get the standard deviation.
```

```python
Skewness is a statistical measure that describes the asymmetry or lack of symmetry in a distribution of data. 
In a symmetrical distribution, the right and left sides are mirror images of each other. 
Skewness quantifies the extent and direction of skew (departure from horizontal symmetry) in a dataset. 
There are two primary types of skewness:

Positive Skewness (Right Skewness): 1. In a positively skewed distribution, the tail on the right side is longer or fatter than the left side. 
                                       The majority of the data points are concentrated on the left side, and the right tail is stretched out.

                                    2. The mean is typically greater than the median in a positively skewed distribution because the presence of larger values in the right tail pulls the mean in that direction.

Negative Skewness (Left Skewness): 1. In a negatively skewed distribution, the tail on the left side is longer or fatter than the right side. 
                                      The bulk of the data points are concentrated on the right side, and the left tail is stretched out.
                                   
                                   2. The mean is usually less than the median in a negatively skewed distribution because the presence of smaller values in the left tail pulls the mean in that direction.


Mathematically, skewness can be measured using different formulas, but a common one is based on the third standardized moment.
```

# Question 9

```python
In a right-skewed distribution, the tail on the right side is longer or fatter than the left side. 
This means that there are some relatively large values pulling the mean in the direction of the longer tail. 
Consequently, in a right-skewed distribution:

The mean will be greater than the median.
This is because the mean is sensitive to extreme values (outliers) in the dataset, and the presence of larger values in the right tail (longer tail) will pull the mean in that direction. On the other hand, the median is less affected by extreme values, as it represents the middle value when the data is ordered.

So, if you encounter a right-skewed distribution, you can expect the mean to be on the right side of the median. 
The relationship between the mean and median is a useful indicator of the skewness of the data distribution.
```

```python
1. Covariance : Covariance measures the extent to which two variables vary together. 
                Specifically, it measures the degree to which two variables deviate from their means in similar ways. 
                Mathematically, the covariance between two variables X and Y is calculated as follows:

                          cov(X,Y) = E[(X - E[X])(Y - E[Y])]
                
                The value of the covariance can range from negative infinity to positive infinity. A positive covariance indicates that the two variables tend to move in the same direction, while a negative covariance indicates that the two variables tend to move in opposite directions. A covariance of zero indicates that the two variables are uncorrelated.

2. Correlation : Correlation, on the other hand, is a standardized measure of the linear relationship between two variables. 
                 Correlation takes on values between -1 and 1, where a correlation of 1 indicates a perfect positive relationship, a correlation of -1 indicates a perfect negative relationship, and a correlation of 0 indicates no linear relationship.
                 The correlation coefficient between two variables X and Y is calculated as follows:

                          corr(X,Y) = cov(X,Y) / (SD[X]*SD[Y])
 
                 Correlation is a more commonly used measure than covariance, because it is standardized and easier to interpret. Correlation can also be used to compare the strength of the relationship between different pairs of variables on a common scale.
```

# Question 12

```python
For a normal distribution, the measures of central tendency (mean, median, and mode) are related in a specific way:

Mean (): The mean of a normal distribution is equal to its median and mode.
          
          In a perfectly symmetrical normal distribution, 
          where the left and right sides of the distribution are mirror images, 
          the mean is located at the center, and it is the point where the distribution is balanced.

Median: The median of a normal distribution is equal to its mean and mode.
        
        In a perfectly symmetrical normal distribution, 
        the median is the same as the mean because the distribution is balanced.

Mode: The mode of a normal distribution is equal to its mean and median.
      
      In a normal distribution, there is only one mode, 
      and it occurs at the peak of the symmetric bell-shaped curve.

In summary, for a normal distribution:

      Mean = Median = Mode

This relationship holds true for a perfectly normal distribution. 
However, in real-world situations, data may exhibit slight deviations from normality,and the equality of these measures may not be exact. 
In practice, analysts often use the mean as the primary measure of central tendency for normal distributions due to its mathematical convenience and efficiency in various statistical analyses.
```

```python
Outliers can significantly impact measures of central tendency and dispersion, 
often leading to biased or misleading results. 
Here's how outliers affect these measures:

* Measures of Central Tendency (Mean, Median, Mode):

   Mean: Outliers have a pronounced effect on the mean because it is sensitive to extreme values. 
         A single outlier can pull the mean in its direction, making it larger or smaller than expected.
   
   Median: The median is less affected by outliers because it is based on the middle value of the ordered data. 
           It provides a more robust measure of central tendency in the presence of extreme values.
   
   Mode: Like the median, the mode is less influenced by outliers, as it represents the most frequently occurring value.

   Example: Consider the dataset: [10, 15, 20, 25, 30, 200]

            Mean: (10 + 15 + 20 + 25 + 30 + 200) / 6 = 58.33
             
            Median: 25 (the middle value)
            
            Mode: No mode

   In this example, the outlier (200) significantly affects the mean, pulling it higher. The median, however, remains relatively unaffected.

* Measures of Dispersion (Range, Variance, Standard Deviation):
 
   Range: Outliers can greatly influence the range, as it is based on the difference between the maximum and minimum values.
 
   Variance and Standard Deviation: These measures are sensitive to the distance between individual data points and the mean. Outliers increase the squared deviations from the mean, leading to inflated values for variance and standard deviation.

   Example: Consider two datasets:

            Dataset A: [5, 10, 15, 20, 25, 200]

            Dataset B: [5, 10, 15, 20, 25, 30]

            Range:

                  Range of A: 200 - 5 = 195
                        
                  Range of B: 30 - 5 = 25

            Variance and Standard Deviation: For both A and B, the presence of the outlier 200 increases the variance and standard deviation.

    In summary, outliers can distort measures of central tendency, especially the mean, and inflate measures of dispersion. 
    Therefore, it's important to identify and handle outliers appropriately to obtain accurate and meaningful summary statistics from a dataset.
```
