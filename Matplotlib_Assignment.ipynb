```python
'''Line Plot:

Visualizes data points as a series of connected lines, suitable for displaying trends over a continuous variable.'''

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

```python
'''Scatter Plot:

Represents individual data points as markers on a 2D plane, useful for showing the relationship between two variables'''

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
'''Histogram:

Displays the distribution of a single variable, showing the frequency of different values within specified bins.'''


import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5]

plt.hist(data, bins=5, edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
'''Bar Plot:

Represents categorical data with rectangular bars, allowing for visual comparison between different categories.'''

import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

```python
'''Pie Chart:

Illustrates the proportion of each component in a whole, often used to represent percentages.'''

import matplotlib.pyplot as plt

labels = ['Category A', 'Category B', 'Category C']
sizes = [30, 40, 30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.show()
```

```python
'''A scatter plot is a type of data visualization that displays individual data points on a two-dimensional plane. 
Each point on the plot represents the values of two variables, allowing for the observation of patterns, relationships, or trends between them. 
Scatter plots are particularly useful for identifying correlations and outliers in the data.'''

import numpy as np
import matplotlib.pyplot as plt

# Generate random data for x and y
np.random.seed(42)  # Setting seed for reproducibility
num_points = 50
x = np.random.rand(num_points)
y = 2 * x + 1 + 0.1 * np.random.randn(num_points)  # y = 2x + 1 + random noise

# Plot the scatter plot
plt.scatter(x, y, color='blue', label='Data Points')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np

# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])


'''The subplot() function in Matplotlib is used to create multiple plots in the same figure.
This function allows us to divide the figure into a grid of rows and columns, and then specify which subplot to plot in each cell of the grid.'''

# Importing Matplotlib
import matplotlib.pyplot as plt

# Create four line plots in one figure using subplot()
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

# Plot 1 : Line 1
axes[0, 0].plot(x1, y1)
axes[0, 0].set_title('Line 1')

# Plot 2 : Line 2
axes[0, 1].plot(x2, y2)
axes[0, 1].set_title('Line 2')

# Plot 3 : Line 3
axes[1, 0].plot(x3, y3)
axes[1, 0].set_title('Line 3')

# Plot 4 : Line 4
axes[1, 1].plot(x4, y4)
axes[1, 1].set_title('Line 4')

# Add a common x-axis label and y-axis label
fig.text(0.5, 0.04, 'X-axis', ha='center')
fig.text(0.04, 0.5, 'Y-axis', va='center', rotation='vertical')

# Display the plot
plt.show()
```

```python
'''A bar plot, also known as a bar chart or bar graph, is a graphical representation of categorical data with rectangular bars. 
Each bar's length or height corresponds to the value it represents, making it easy to compare different categories. 
Bar plots are commonly used to visualize the distribution of categorical variables, compare values between categories, and show the relationships or trends within the data.'''

'''-> Here are some specific reasons why we should use bar plot:'''

'''Comparison of Categorical Data: Bar plots are excellent for visually comparing the quantities or frequencies of different categories. 
The length or height of each bar directly represents the value it represents, making it easy to identify the largest and smallest values.

Distribution of Data: Bar plots provide a clear visual representation of the distribution of categorical data. 
They are useful for understanding the spread and concentration of values across different categories.

Emphasis on Differences: Bar plots emphasize the differences between categories, making it easy to identify which categories have higher or lower values. 
This is particularly useful for highlighting trends or variations.

Categorical Data Trends Over Time: Bar plots can be used to show how the values of different categories change over time. 
Stacked bar plots or grouped bar plots can be employed to visualize trends in multi-category datasets.

Comparison Across Subgroups:Bar plots can be used to compare values across subgroups within each category. 
Grouped bar plots or stacked bar plots are effective for representing data with multiple dimensions.'''
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Plotting Normal Bar plot
import matplotlib.pyplot as plt
plt.bar(company,profit)
plt.xlabel('Company Name')
plt.ylabel('Profit')
plt.title('Profit Comparison for various Companies')
plt.show()
```

```python
# Plotting Horizontal Bar plot
import matplotlib.pyplot as plt
plt.barh(company,profit)
plt.xlabel('Profit')
plt.ylabel('Company Name')
plt.title('Profit Comparison for various Companies')
plt.show()
```

```python
'''A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of the distribution of a dataset. It displays the median, quartiles, and potential outliers in a compact and informative manner. The key components of a box plot include:

Box: Represents the interquartile range (IQR), indicating the central 50% of the data. 
The box is divided into the first quartile (Q1) and the third quartile (Q3), with the median (Q2) marked inside the box.

Whiskers: Lines extending from the box that indicate the range of the data. 
Typically, whiskers extend to the minimum and maximum values within a certain range (often 1.5 times the IQR). Points beyond the whiskers are considered potential outliers.

Outliers: Individual data points that fall outside the whiskers and are displayed as dots.

Box plots are used for several reasons:

Summary of Distribution: Box plots provide a concise summary of the distribution of a dataset, highlighting central tendency, spread, and potential outliers.

Comparison of Groups: Box plots are useful for comparing the distribution of multiple groups or categories side by side, allowing for visual comparisons.

Identification of Outliers: Outliers, which can skew the interpretation of data, are visually identified on the box plot.

Detection of Skewness: Box plots help in identifying whether the distribution is symmetric or skewed.

Visualization of Quartiles: Quartiles provide insights into the range in which the central 50% of the data falls.'''
```

```python
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
# Importing Matplotlib
import matplotlib.pyplot as plt

# Combine the data into a list
data = [box1, box2]

# Create the box plot
fig, ax = plt.subplots()
ax.boxplot(data,patch_artist=True, boxprops=dict(facecolor='lightblue'))

# Add labels and title
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')
ax.set_title('Box Plot of Two Datasets')

# Display the plot
plt.show()
```
