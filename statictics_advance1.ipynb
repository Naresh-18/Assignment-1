```python
Probability density function is a function that provides the likelihood that the value of a random variable will fall between a certain range of values. 
We use the probability density function in the case of continuous random variables. For discrete random variables, we use the probability mass function which is analogous to the probability density function.

The PDF of a continuous random variable is defined as the derivative of its cumulative distribution function (CDF). 
The CDF gives the probability that a random variable takes on a value less than or equal to a given value.

The PDF is used to calculate the probability that a random variable falls within a certain range of values, by integrating the PDF over that range. 
The area under the PDF curve represents the total probability of all possible values of the random variable.

In summary, the PDF provides a way to quantify the probability distribution of a continuous random variable, making it a useful tool in many fields including statistics, physics, and engineering.
```

There are many types of probability distributions, but some of the most common ones are:

1. Normal Distribution (Gaussian Distribution): A symmetric probability distribution that forms a bell-shaped curve. It is completely characterized by its mean () and standard deviation ().
      
                                           

2. Binomial Distribution: Describes the number of successes in a fixed number of independent and identical Bernoulli trials, where each trial has two possible outcomes (success or failure).



3. Poisson Distribution: Models the number of events that occur in a fixed interval of time or space in a process with a constant rate. It is often used to describe rare events.



4. Exponential Distribution: Models the time between events in a Poisson process. It is characterized by its constant rate parameter and exhibits the memoryless property.



5. Uniform Distribution: A probability distribution where all outcomes in a given range are equally likely. The probability density is constant within the range and zero outside the range.

```python
import math

def normal_pdf(x, mean, std_dev):
    """
    Calculates the probability density function (PDF) of a normal distribution
    with a given mean and standard deviation at a given point x.
    """
    variance = std_dev ** 2
    exponent = -((x - mean) ** 2) / (2 * variance)
    coefficient = 1 / math.sqrt(2 * math.pi * variance)
    pdf = coefficient * math.exp(exponent)
    return pdf
```

```python
normal_pdf(x=3,mean=2,std_dev=1.5)
```

# Question 4

```python
Properties of Binomial Distribution:

Number of Trials (n): The binomial distribution is based on a fixed number of independent and identical trials.

Two Outcomes (Success and Failure): Each trial has only two possible outcomes, often denoted as success (S) or failure (F).

Probability of Success (p): The probability of success remains constant for each trial.

Independence: The trials are assumed to be independent, meaning the outcome of one trial does not influence the outcome of another.

Discreteness: The random variable representing the number of successes is discrete, taking on integer values.

Binomial Coefficient: The probability mass function of the binomial distribution involves the binomial coefficient, denoted as 
(n k), which represents the number of ways to choose k successes from n trials.

Examples of Events Where Binomial Distribution Can Be Applied:

1. Coin Tossing:

   Experiment: Tossing a fair coin multiple times.
   
   Outcomes: Success (e.g., getting heads) or failure (getting tails).
    
   Parameters: n trials, p probability of success (0.5 for a fair coin).

2. Quality Control:

   Experiment: Inspecting a batch of items for defects.

   Outcomes: Each item is either defective or non-defective.

   Parameters: n items, p probability of an item being defective.
   
   Example: Determining the probability of finding exactly k defective items in a sample of n items.

The binomial distribution is widely applicable in various real-world scenarios where events have only two possible outcomes, 
and the trials are independent and follow the same probability of success.
```

```python
import math

def poisson_cdf(mean, point):
    """
    Calculates the cumulative distribution function of a 
    Poisson distribution with given mean at a given point
    """
    cdf = 0
    for k in range(point + 1):
        cdf = cdf + math.exp(-mean) * pow(mean, k) / math.factorial(k)
    return cdf
```

```python
result = poisson_cdf(mean=4,point=2)
print(f'Poison cdf for mean = 4 at x = 2 is {round(result,4)}')
```

# Question 10

In a normal distribution, the least frequent data appear in the tails of the distribution, which are the regions that are farthest from the mean. 
The tails of a normal distribution represent the extreme values of the variable being measured, which are less common than the values closer to the mean.


For example, if we have a normal distribution with mean 0 and standard deviation 1, the least frequent data will appear in the tails of the distribution, 
which are the regions beyond 3 standard deviations from the mean. 
This is because the normal distribution is symmetric around the mean, 
and the probability density decreases rapidly as we move away from the mean towards the tails of the distribution. 
The probability of observing a value in the tails of a normal distribution is relatively low compared to the probability of observing a value closer to the mean.
